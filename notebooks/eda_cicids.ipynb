```python
# 1. Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# Options pandas/seaborn
pd.set_option("display.max_columns", 100)
sns.set(style="whitegrid")

# 2. Définir chemin des fichiers
raw_dir = Path("../data/raw")
csv_files = list(raw_dir.glob("*.csv"))

print(f"Nombre de fichiers trouvés: {len(csv_files)}")
for f in csv_files:
    print("-", f.name)

# 3. Charger un échantillon de chaque fichier
dfs = []
for f in csv_files:
    try:
        # Charger seulement 100k lignes max (sinon explosion RAM)
        df_part = pd.read_csv(f, nrows=100000)
        df_part.columns = df_part.columns.str.strip()  # nettoyer colonnes
        dfs.append(df_part)
        print(f"[INFO] Chargé {f.name} → {df_part.shape}")
    except Exception as e:
        print(f"[ERREUR] {f.name}: {e}")

# Concaténer les samples
df = pd.concat(dfs, ignore_index=True)
print("\nShape global après sampling:", df.shape)

# 4. Aperçu
df.head()
```

```python
df.describe().T
```

```python
print(df["Label"].value_counts())

sns.countplot(y="Label", data=df, order=df["Label"].value_counts().index)
plt.title("Répartition des classes (CICIDS, sample)")
plt.show()
```

```python
df["binary_label"] = df["Label"].apply(lambda x: "BENIGN" if x=="BENIGN" else "ATTACK")
sns.countplot(x="binary_label", data=df)
plt.title("Normal vs Attaque (CICIDS sample)")
plt.show()
```
